```python
from bs4 import BeautifulSoup
import requests
from konlpy.tag import Twitter
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os
```

```python
# 실행 x
def make_wordcloud(word_count):
    twitter = Twitter()
 
    sentences_tag = []
    #형태소 분석하여 리스트에 넣기
    word_text = os.listdir('/content/')
    for file in word_text:
      if '금융통화위원회' in file :
        word_text =open('/content/' + file, 'r', encoding='utf-8').read()
        break
    
    for sentence in word_text:
        morph = twitter.pos(sentence)
        sentences_tag.append(morph)
        print(morph)
        print('-' * 30)
 
    print(sentences_tag)
    print('\n' * 3)
 
    noun_adj_list = []
    #명사와 형용사만 구분하여 이스트에 넣기
    for sentence1 in sentences_tag:
        for word, tag in sentence1:
            if tag in ['Noun', 'Adjective']:
                noun_adj_list.append(word)
 
    #형태소별 count
    counts = Counter(noun_adj_list)
    tags = counts.most_common(word_count)
 
    #wordCloud생성
    #한글꺠지는 문제 해결하기위해 font_path 지정
    wc = WordCloud(font_path='NanumSquareL.ttf', background_color='white', width=800, height=600)
    #wc.generate_from_frequencies()
    cloud_1 = wc.generate_from_frequencies(dict(tags))
    plt.figure(figsize=(10, 8))
    plt.axis('off')
    plt.imshow(cloud_1)
    plt.show()

def temp_12(word_text):
    twitter = Twitter()
 
    sentences_tag = []
    #형태소 분석하여 리스트에 넣기
    for sentence in word_text:
        morph = twitter.pos(sentence)
        sentences_tag.append(morph)

# 폴더 경로 수정
make_wordcloud(30)
```

```python
from collections import Counter #데이터의 개수를 정리할 수 있는 모듈
from konlpy.tag import Kkma #형태소 분석기 호출
from konlpy.utils import pprint #유니코드 문자 출력
import pandas as pd #결과값을 데이터프레임 객체로 저장하기 위해 이용
from wordcloud import WordCloud
import matplotlib.pyplot as plt #그래프 그리기
from PIL import Image
import numpy as np

word_text = os.listdir('/content/')
for file in word_text:
  if '금융통화위원회' in file :
    word_text =open('/content/' + file, 'r', encoding='utf-8').read()
    break

kkma = Kkma()

nouns_list=[] #nouns_list생성
#for item in word_text: #[ ]안의 숫자에 따라 분석 단어 개수 변화

sentence_list = kkma.sentences(word_text)
for sentence in sentence_list:

    nouns = kkma.pos(sentence) #sentence의 형태소 분석

    for pos in nouns:

        if pos[1] == 'NNG' or pos[1] == 'NNP': #일반 명사와 고유 명사일 경우

            nouns_list+=[pos[0]] #nouns_list에 추가

for word in ignore:
    if word in nouns_list:
        nouns_list.remove(word) #위에서 정한 불필요한 단어 및 표현 리스트에 속한 단어 및 표현을 제거
        
count = Counter(nouns_list) #list 내 항목명과 항목별 개수
```

```python
# 30 개까지 추출
count.most_common()[:30]

num_list=[] #nouns_list생성

for temp,cnt in count.most_common()[:30]:
    for i in range(1,cnt+1):
      num_list.append(temp)

num_cnt = Counter(num_list)
```

```python
im = Image.open('sharlock.png') # 이미지 파일 읽어오기
mask_arr = np.array(im) # 픽셀 값 배열 형태 변환

#font_path = '‪NanumSquareL.ttf' #글꼴 : 휴먼둥근헤드라인체
# 색맵 추천 
wc = WordCloud(background_color ='white', colormap='autumn',random_state = 43, mask = mask_arr,
                      prefer_horizontal = True, font_path='NanumSquareL.ttf', width=800, height=600)

wordcloud_1 = wc.generate_from_frequencies(num_cnt) #단어 출현 빈도에 따라 크기 변화 정렬 count 전체, num_cnt= 30개까지 

array = wordcloud_1.to_array()

fig = plt.figure(figsize=(8, 6))

plt.title("Word Frequency", size = 13)
plt.imshow(array, interpolation="bilinear")
plt.axis('off')
plt.savefig('./sin.png')
plt.show() #word cloud 도출
```

```python
from konlpy.tag import Kkma
kkma = Kkma()

import sys
import os
import re
#sys.path.append(os.path.dirname('PyKoSpacing/'))
from pykospacing import spacing

def preprocessing(review):
    total_review = ''
    #인풋리뷰
    for idx in range(len(review)):
        r = review[idx]
        #하나의 리뷰에서 문장 단위로 자르기
        for sentence in kkma.sentences(r):
            sentence = re.sub('([a-zA-Z])','',sentence)
            sentence = re.sub('[ㄱ-ㅎㅏ-ㅣ]+','',sentence)
            sentence = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',sentence)
            if len(sentence) == 0:
                continue
            if len(sentence) < 198:
                sentence = spacing(sentence)
            sentence += '. '
            total_review += sentence
    return total_review
```
